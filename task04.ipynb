```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("twitter_training.csv", header=None)
df.columns = ["id", "entity", "sentiment", "tweet"]
df.head()
```

```python
df = df.dropna(subset=["tweet"])
df["sentiment"].value_counts()

```

```python
sentiment_counts = df["sentiment"].value_counts()
sentiment_counts
```

```python
sentiment_counts.plot(kind="bar")
plt.title("Overall Sentiment Distribution on Twitter Data")
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")
plt.show()
```

```python
amazon_df = df[df["entity"] == "Amazon"]
amazon_sentiment = amazon_df["sentiment"].value_counts()

amazon_sentiment.plot(kind="bar")
plt.title("Sentiment Analysis for Amazon")
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")
plt.show()

```
